```python
import cv2  # Importing OpenCV
import numpy as np  # Importing Numpy as np
from matplotlib import pyplot as plt  # Importing pyplot from matplotlib as plt

# This is our imshow function
def imshow(title="", image=None, size = 10):
  w, h = image.shape[0], image.shape[1]
  aspect_ratio = w/h
  plt.figure(figsize=(size * aspect_ratio, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
image = cv2.imread("/content/drive/MyDrive/Sample Photos/Cars.jpg") # Loading our image
```

```python
imshow("Cars", image)
```

```python
height, width = image.shape[:2]   # Extracting the height and witdh of the image
quarter_height, quarter_width = height/4, width/4 # We will be shifting the image by quarter of it's width and height

T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]]) # Now, we are creating our transaltion matrix

# Then we use the cv2.warpAffine() function in which we give the input image, the translation matrix and it's width and height
translated_image = cv2.warpAffine(image, T, (width, height))

# Now, we are displaying the original and translated image
imshow("Original Image", image, size = 5) 
imshow("Translated Image", translated_image, size = 5)
```

```python
# Now, let's try to rotate our image 
# We will take the centre coordinates by dividing the height and width of the image by 2
centre = (height/2, width/2)

# Now, we form our rotion matrix
# We will be rotating our image by 90 degrees
# We will be setting the scale to 0.5
rotation_matrix = cv2.getRotationMatrix2D(centre, 90, 0.5)

# Finally we use the cv2.warpAffine() function with the parameters as used above
# We will just be passing the rotation matrix instead of the translation matrix
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
```

```python
# Finally, we show our rotated image
imshow("Rotated Image", rotated_image)
```

```python
# We can also use the cv2.transpose function which is considered better as it does not leave any black border around the image
rotated_image = cv2.transpose(image)

imshow("Original Image", image, size = 5)
imshow("Rotated Image", rotated_image, size = 5)
```

```python
# We can also use the cv2.flip() function to flip our image
# This also be understood as the mirror image of the original image
flipped = cv2.flip(image, 1)

imshow("Flipped Image", flipped, size = 5)
```
