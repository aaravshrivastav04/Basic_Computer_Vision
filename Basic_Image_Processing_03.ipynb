```python
import cv2 # Importing OpenCV
import numpy as np # Importing Numpy as np
from matplotlib import pyplot as plt # Importing pyplot from matplotlib as plt

def imshow(title="", image=None, size = 10):
  h, w = image.shape[:2]
  aspect_ratio = w/h
  plt.figure(figsize=(size * aspect_ratio, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
img = cv2.imread("/content/drive/MyDrive/Sample Photos/Image.jpg")
imshow("Image", img, size = 5)

cv2.rectangle(img, (290, 70), (500, 260), (255, 255, 255), 2)

cv2.putText(img, "Happy Dog!", (270, 300), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2)
imshow("Image", img, size = 5)
```

```python
# Now, let;s draw circles

# For that, let's first create a black canvas on which we can test our opencv drawing functions

canvas = np.zeros((500, 500, 3), np.uint8)

# Now, let's draw a circle of radius  of radius 10 units at the centre

cv2.circle(canvas, (250, 250), 100, (255, 0, 0), 1)
imshow("Circle", canvas)
```

```python
# We can also draw a filled circle

canvas = np.zeros((500, 500, 3), np.uint8)

# We simply need to add the minus sign to the thickness

cv2.circle(canvas, (250, 250), 100, (255, 0, 0), -1)
imshow("Circle Filled", canvas)
```

```python
# We draw lines using cv2.line() function

canvas = np.zeros((500, 500, 3), np.uint8)

cv2.line(canvas, (0, 0), (500, 500), (255, 255, 255), 2)

imshow("Line", canvas)
```

```python
# Now, let's draw a polygon

canvas = np.zeros((500, 500, 3), np.uint8)

pts = np.array([[50, 50], [100, 100], [175, 480], [254, 190]], np.int32)

pts = pts.reshape(-1, 1, 2)

cv2.polylines(canvas, [pts], True, (0, 255, 0), 2)

imshow("Polygon", canvas)
```
